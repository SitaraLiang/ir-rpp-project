# Student test and Adapted HSD Test for Web Subtopic

The paper specifies the comparaison of four metrics:

- MAP-IA: No rank cutoff.

- ERR-IA@20: Rank cutoff of 20.

- ST-Recall(strec)@20: Rank cutoff of 20.

- ST-RPP: rpp metric, applied per subtopic and averaged.
